```python
names = ["p_id","no_times_pregnantglucose_concentration","blood_pressure","skin_fold_thickness","serum_insulin","bmi","diabetes pedigree","age","diabetes"]
correlations = data.corr()
plt.rcParams["figure.figsize"] = (20,3)
# plot correlation matrix
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```

```python
array = data.values
X = array[:,0:9]
Y = array[:,9]

scaler = MinMaxScaler(feature_range=(0, 1)).fit(X)
X_scaled = scaler.transform(X)

X_scaled
```

```python
#split into training an test dataset
x_train, x_test, y_train, y_test = train_test_split(X_scaled, Y, random_state=1)
```

# Logistic regresion

```python
# Create a Logistic Regression Object, perform Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
log_reg.score(x_test,y_test)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure()
sns.heatmap(cm , annot =True, linewidth=0.8,fmt=".1f")
plt.title('Logistic Regression')
plt.show()
```

# K-fold Cross Validation


```python
num_folds = 10
kfold = KFold(n_splits=num_folds)
model = LogisticRegression()
results = cross_val_score(model, X_scaled, Y, cv=kfold)
print(f"accurracy: {results.mean()*100}%,{results.std()*100}%")
```

# Linear regresion
